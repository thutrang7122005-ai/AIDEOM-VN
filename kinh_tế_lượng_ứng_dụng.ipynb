```
from linearmodels.panel import PanelOLS
import statsmodels.api as sm
```

```
panel_fe = panel_clean.set_index(
    ['id','year']
)
```

```
X = panel_fe[
    [
        'credit',
        'assets',
        'labor'
    ]
]

X = sm.add_constant(X)

y = panel_fe['profit']

model = PanelOLS(
    y,
    X,
    entity_effects=True
)

result = model.fit(
    cov_type='robust'
)

print(result)
```

```
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()

vif["Variable"] = X.columns

vif["VIF"] = [
    variance_inflation_factor(
        X.values,
        i
    )
    for i in range(X.shape[1])
]

print(vif)
```

```
with open(
    '/content/drive/MyDrive/SME_Project/fixed_effects_result.txt',
    'w'
) as f:
    f.write(str(result.summary))
```

```
import pandas as pd

# 1. Thống kê mô tả
desc = panel_clean[['profit','credit','assets','labor']].describe()

# 2. Ma trận tương quan
corr = panel_clean[['profit','credit','assets','labor']].corr()

# 3. Cấu trúc panel
panel_structure = (
    panel_clean.groupby('id')['year']
    .count()
    .value_counts()
    .sort_index()
    .reset_index()
)

panel_structure.columns = [
    'Number_of_Years_Appearing',
    'Number_of_Firms'
]

# 4. Thông tin tổng quát mẫu
sample_info = pd.DataFrame({
    'Indicator': [
        'Total Observations',
        'Unique Firms',
        'Year 2011 Observations',
        'Year 2013 Observations',
        'Year 2015 Observations'
    ],
    'Value': [
        len(panel_clean),
        panel_clean['id'].nunique(),
        (panel_clean['year']==2011).sum(),
        (panel_clean['year']==2013).sum(),
        (panel_clean['year']==2015).sum()
    ]
})

# 5. Kết quả hồi quy FE
fe_result = pd.DataFrame({
    'Regression_Output': str(result.summary).split('\n')
})

# 6. Xuất Excel
with pd.ExcelWriter(
    '/content/drive/MyDrive/SME_Project/Research_Results.xlsx'
) as writer:

    sample_info.to_excel(
        writer,
        sheet_name='Sample_Info',
        index=False
    )

    panel_structure.to_excel(
        writer,
        sheet_name='Panel_Structure',
        index=False
    )

    desc.to_excel(
        writer,
        sheet_name='Descriptive_Statistics'
    )

    corr.to_excel(
        writer,
        sheet_name='Correlation_Matrix'
    )

    fe_result.to_excel(
        writer,
        sheet_name='Fixed_Effects_Result',
        index=False
    )

print("Xuất file thành công!")
```

```
import numpy as np

panel_clean['ln_profit'] = np.log1p(panel_clean['profit'])

panel_clean['ln_credit'] = np.log1p(panel_clean['credit'])

panel_clean['ln_assets'] = np.log1p(panel_clean['assets'])

panel_clean['ln_labor'] = np.log1p(panel_clean['labor'])
```

```
panel_log = panel_clean.set_index(
    ['id','year']
)
```

```
from linearmodels.panel import PanelOLS
import statsmodels.api as sm

X = panel_log[
    [
        'ln_credit',
        'ln_assets',
        'ln_labor'
    ]
]

X = sm.add_constant(X)

y = panel_log['ln_profit']

model = PanelOLS(
    y,
    X,
    entity_effects=True
)

result_log = model.fit(
    cov_type='robust'
)

print(result_log.summary)
```

```
coef_table = pd.DataFrame({
    'Variable': result_log.params.index,
    'Coefficient': result_log.params.values,
    'P_value': result_log.pvalues.values
})

coef_table
```

```
coef_table.to_excel(
    '/content/drive/MyDrive/SME_Project/Regression_Log_Result.xlsx',
    index=False
)
```

```
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_vif = panel_clean[
    ['ln_credit',
     'ln_assets',
     'ln_labor']
]

vif = pd.DataFrame()

vif["Variable"] = X_vif.columns

vif["VIF"] = [
    variance_inflation_factor(
        X_vif.values,
        i
    )
    for i in range(X_vif.shape[1])
]

print(vif)
```

```
import pandas as pd

coef_table = pd.DataFrame({
    'Coefficient': result_log.params,
    'Std_Error': result_log.std_errors,
    'T_Statistic': result_log.tstats,
    'P_Value': result_log.pvalues,
    'CI_Lower': result_log.conf_int().iloc[:,0],
    'CI_Upper': result_log.conf_int().iloc[:,1]
})

coef_table
```

```
model_info = pd.DataFrame({
    'Statistic': [
        'Number of Observations',
        'Number of Firms',
        'R-squared Within',
        'R-squared Between',
        'R-squared Overall',
        'F-statistic Robust',
        'Prob(F-statistic)',
        'Poolability Test',
        'Poolability P-value'
    ],
    'Value': [
        result_log.nobs,
        panel_clean['id'].nunique(),
        result_log.rsquared_within,
        result_log.rsquared_between,
        result_log.rsquared_overall,
        result_log.f_statistic.stat,
        result_log.f_statistic.pval,
        result_log.f_pooled.stat,
        result_log.f_pooled.pval
    ]
})
```

```
with pd.ExcelWriter(
    '/content/drive/MyDrive/SME_Project/Log_FE_Results.xlsx'
) as writer:

    coef_table.to_excel(
        writer,
        sheet_name='Regression_Coefficients'
    )

    model_info.to_excel(
        writer,
        sheet_name='Model_Information',
        index=False
    )

    desc.to_excel(
        writer,
        sheet_name='Descriptive_Statistics'
    )

    corr.to_excel(
        writer,
        sheet_name='Correlation_Matrix'
    )

print("Đã xuất thành công!")
```
